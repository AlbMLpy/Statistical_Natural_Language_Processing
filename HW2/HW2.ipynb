## 2.1 Methodology 

The main problem I tried to solve is **Text Detoxification**, meaning that we want to automatically *identify* 'Bad' words/phrases that are swears and *switch them* with more neutral counterparts. This problem can be considered as style transfering problem.

Some examples:

1. "This **idiot** at Buckingham Palace has stolen my wallet" $\rightarrow$  "This **unclever person** at Buckingham Palace has stolen my wallet"

2. "Как же твоя *отвратительная*(?) игра меня **за....а**" $\rightarrow$ "Как же твоя *плохая*(?) игра меня **утомляет**"

As we can see from the examples above this problem has many solutions(synonimy) and besides it is not pretty obvious what words should we define as "bad".

*The main idea is that we have to save content/context and substitute only bad style.*

Модель и предобработанные эмбеддинги

In this project, I tried and compared several models like:
1. Duplicate - save a sentence as it is without any substitutions;
2. TDModel(Pointwise editing model) - based on:
    - multilanguage pretrained RoBERTa (https://huggingface.co/unitary/multilingual-toxic-xlm-roberta);
    - pretrained word embeddings;
    - switching the most toxic words with their more neutral equivalents; 
3. condBert - (https://github.com/skoltech-nlp/rudetoxifier);

Some important references:
- http://www.dialog-21.ru/media/5366/dementyevaplusetal.pdf
- https://arxiv.org/pdf/2105.09052.pdf

To tackle this problem one would go through these steps:

1. **Data preprocessing**: 
- tokenization(word level);
- make all the words lower case;

2. **Model Definition**(training=optional, depends on the model): 
- word-toxicity dictionaty from pretrained classifier;
- represent all the words as pretrained embeddings;
- for the most toxic words find the most similar words in terms of embeddings;

3. **Model evaluation**: 
- STA(Style transfer accuracy): used method that is based on pretrained classifier model.
- chF(character n-gram F-score): used nltk.translate.chrf_score.sentence_chrf and averaged the results.

4. **Send the results**: the .txt -> .zip files to CodaLab system.

*Some words about why I decided to move this particular way:*

*First of all*, The model TDModel looks like pretty straightforward and we can use the pretrained models that allows for faster experiments conduction.

*Secondly*, it is really interesting to find out if this model can beat the baseline.

## 3.3 Preprocessing 

```python
import pandas as pd
from transformers import AutoTokenizer

TOKENIZATION_TYPE = 'sentencepiece'
tokenizer = AutoTokenizer.from_pretrained("trained_roberta/")

def is_word_start(token):
    if TOKENIZATION_TYPE == 'sentencepiece':
        return token.startswith('▁')
    if TOKENIZATION_TYPE == 'bert':
        return not token.startswith('##')
    raise ValueError("Unknown tokenization type")


def normalize(sentence, max_tokens_per_word=20):
    sentence = ''.join(map(lambda c: c if c.isalpha() else ' ', sentence.lower()))
    ids = tokenizer(sentence)['input_ids']
    tokens = tokenizer.convert_ids_to_tokens(ids)[1:-1]
    
    result = []
    num_continuation_tokens = 0
    for token in tokens:
        if not is_word_start(token):
            num_continuation_tokens += 1
            if num_continuation_tokens < max_tokens_per_word:
                result.append(token.lstrip('#▁'))
        else:
            num_continuation_tokens = 0
            result.extend([' ', token.lstrip('▁#')])
    
    return ''.join(result).strip()
```

## 3.4 My method of text processing

```python
import numpy as np
import torch
import gensim
from torch.nn.utils.rnn import pad_sequence
from torch import softmax, sigmoid
from transformers import (
    AutoModelForSequenceClassification,
    BertTokenizer,
    BertForSequenceClassification,
)
from pymystem3 import Mystem
from sklearn.neighbors import KDTree
from nltk.translate.chrf_score import sentence_chrf

from tqdm.auto import tqdm
```

```python
class TDModel:
    
    def __init__(self, path_model, path_embeddings):
        # Define constant objects:
        self._toxic_class = -1
        self._token_type = "sentencepiece"
        
        self.alphabet = list(map(chr, range(ord('а'), ord('я') + 1)))
        self.alphabet.extend(list(map(str.upper, self.alphabet)))
        self.alphabet = set(self.alphabet)

        self.normalizer = normalize
        self.stemmer = Mystem()
        
        # To find/classify toxic words in a sentence:
        self.tokenizer = AutoTokenizer.from_pretrained(path_model)
        self.model = AutoModelForSequenceClassification.from_pretrained(path_model).cuda()

        # To find similar words based on precalculated embeddings:
        temp = np.load(path_embeddings, allow_pickle=True)
        self.emb_vec = temp["vectors"]
        self.emb_voc = temp["voc"].item()


    def _logits_to_toxic_probas(self, logits):
        
        if logits.shape[-1] > 1:
            activation = lambda x: softmax(x, -1)
        else:
            activation = sigmoid
        return activation(logits)[:, self._toxic_class].cpu().detach().numpy()


    def _iterate_batches(self, data, batch_size=40):
        
        batch = []
        for x in data:
            batch.append(x)
            if len(batch) >= batch_size:
                yield batch
                batch = []
        if len(batch) > 0:
            yield batch


    def  _predict_toxicity(self, sentences, batch_size=5, threshold=0.5, return_scores=False, verbose=True, device='cuda'):
        
        results = []
        tqdm_fn = tqdm if verbose else lambda x, total: x
        for batch in tqdm_fn(self._iterate_batches(sentences, batch_size), total=np.ceil(len(sentences) / batch_size)):
            normalized = [self.normalizer(sent, max_tokens_per_word=5) for sent in batch]
            tokenized = self.tokenizer(normalized, return_tensors='pt', padding=True, max_length=512, truncation=True)
        
            logits = self.model.to(device)(**{key: val.to(device) for key, val in tokenized.items()}).logits
            preds = self._logits_to_toxic_probas(logits)
            if not return_scores:
                preds = preds >= threshold
            results.extend(preds)
        return results


    def _get_w2v_indicies(self, a):
        
        res = []
        if isinstance(a, str):
            a = a.split()
        for w in a:
            if w in self.emb_voc:
                res.append(self.emb_voc[w])
            else:
                lemma = self.stemmer.lemmatize(w)[0]
                res.append(self.emb_voc.get(lemma, None))
        return res


    def _calc_embs(self, words):
        
        words = ' '.join(map(self.normalizer, words))
        inds = self._get_w2v_indicies(words)
        return [None if i is None else self.emb_vec[i] for i in inds]


    def _find_closest_nontoxic(self, word, threshold=0.5, allow_self=False):
        
        if self.toxicity.get(word, 1.0) <= threshold:
            return word
    
        if word not in self.toxicity and word not in embs_voc:
            return None
    
        threshold = min(self.toxicity.get(word, threshold), threshold)
        word = self.normalizer(word)
        word_emb = self._calc_embs([word])
        if word_emb is None or word_emb[0] is None:
            return None
    
        for i in self.embs_tree.query(word_emb)[1][0]:
            other_word = self.embs_id2voc[self.nontoxic_emb_inds[i]]
            if (other_word != word or allow_self) and self.toxicity.get(other_word, 1.0) <= threshold:
                return other_word
        return None


    def _get_toxicity(self, texts):
        
        words = set()
        for text in texts:
            words.update(text.split())
            #words = sorted(words)

        with torch.inference_mode():
            word_toxicities = self._predict_toxicity(words, batch_size=100, return_scores=True)
    
        return dict(zip(words, word_toxicities))
   

    def _prepare_embeddings_ids(self):
        embs_vectors_normed = self.emb_vec / np.linalg.norm(self.emb_vec, axis=1, keepdims=True)
        embs_voc_by_id = [None for i in range(len(self.emb_vec))]
        for word, idx in self.emb_voc.items():
            if embs_voc_by_id[idx] is None:
                embs_voc_by_id[idx] = word
        return embs_voc_by_id


    def _detox(self, line):
        
        words = self.normalizer(line).split()
        fixed_words = [self._find_closest_nontoxic(word, allow_self=True) or '' for word in words]
        return ' '.join(fixed_words)


    def detoxicate(self, sentences_list):

        # calculate word:toxicity mapping
        self.toxicity = self._get_toxicity(sentences_list)

        # prepare structures for embeddings
        self.embs_id2voc = self._prepare_embeddings_ids()
        embs_vectors_normed = self.emb_vec / np.linalg.norm(self.emb_vec, axis=1, keepdims=True)
        self.nontoxic_emb_inds = [ind for word, ind in self.emb_voc.items() if self.toxicity.get(word, 1.0) <= 0.5]
        embs_vectors_normed_nontoxic = embs_vectors_normed[self.nontoxic_emb_inds]
        self.embs_tree = KDTree(embs_vectors_normed_nontoxic, leaf_size=20)

        return list(map(self._detox, tqdm(sentences_list)))
```

```python
_sta_tokenizer = BertTokenizer.from_pretrained('SkolkovoInstitute/russian_toxicity_classifier')
_sta_model = BertForSequenceClassification.from_pretrained('SkolkovoInstitute/russian_toxicity_classifier')

def sta_score(batch_size, sentences_list):
    
    tokenizer = _sta_tokenizer
    model = _sta_model

    results = []
    for i in tqdm(range(0, len(sentences_list), batch_size)):
        batch = tokenizer(sentences_list[i:i + batch_size], return_tensors='pt', padding=True)
        result = model(**batch)['logits'].argmax(1).float().data.tolist()
        results.extend([1 - item for item in result])

    return results


def chF_score(sentences_list_pred, sentences_list_manual):
    
    pr = sentences_list_pred
    tr = sentences_list_manual
    chF = [sentence_chrf(p, t) for p, t in zip(pr, tr) if (len(p.split()) > 2) and (len(t.split()) > 2)]
    return np.mean(chF)
```
